```python
def convert_to_numeric_and_date(df, crypto):
    num=['Open','High','Low','Close','Volume','Market_Cap']

    df2 = df.copy()
    for n in num:
        for i in range(len(df2)):
            col = crypto + '_' + n
            if type(df2[col][i]) == str:
                df2[col][i]=df2[col][i].replace(',','')
                df2[col][i]=pd.to_numeric(df2[col][i])
                df2[col][i] = int(df2[col][i]) #convert numerical to integer in order to be able to apply statsmodels timeseries seasonal decompose

        col = crypto + '_Date'
    try:
        df2[col] = pd.to_datetime(df2[col])
        #1==1
        #df2['Date'] = pd.to_datetime(df2['Date'])
    except:
        pass
    return(df2)
```

Ajout du numéro du jour de la semaine Nous allons le faire comme suit :

    Ajout d'une colonne ayant comme valeur : cos(2 * PI * num_jour / 7)
    Ajout d'une colonne ayant comme valeur : sin(2 * PI * num_jour / 7) Nous utiliserons cette nomenclature :
    1 : Lundi => cos(2 * PI * 1 / 7) & sin(2 * PI * 1 / 7)
    2 : Mardi => cos(2 * PI * 2 / 7) & sin(2 * PI * 2 / 7)
    3 : Mercredi => cos(2 * PI * 3 / 7) & sin(2 * PI * 3 / 7)
    4 : Jeudi => cos(2 * PI * 4 / 7) & sin(2 * PI * 4 / 7)
    5 : Vendredi => cos(2 * PI * 5 / 7) & sin(2 * PI * 5 / 7)
    6 : Samedi => cos(2 * PI * 6 / 7) & sin(2 * PI * 6 / 7)
    7 : Dimanche => cos(2 * PI * 7 / 7) & sin(2 * PI * 7 / 7)


Ajout du numéro du mois Nous allons le faire comme suit :

    Ajout d'une colonne ayant comme valeur : cos(2 * PI * num_mois / 12)
    Ajout d'une colonne ayant comme valeur : sin(2 * PI * num_mois / 12) Nous utiliserons cette nomenclature :
    1 : Janvier => cos(2 * PI * 1 / 7) & sin(2 * PI * 1 / 12)
    2 : Février => cos(2 * PI * 2 / 7) & sin(2 * PI * 2 / 12)
    3 : Mars => cos(2 * PI * 3 / 7) & sin(2 * PI * 3 / 12)
    4 : Avril => cos(2 * PI * 4 / 7) & sin(2 * PI * 4 / 12)
    5 : Mai => cos(2 * PI * 5 / 7) & sin(2 * PI * 5 / 12)
    6 : Juin => cos(2 * PI * 6 / 7) & sin(2 * PI * 6 / 12)
    7 : Juillet => cos(2 * PI * 7 / 7) & sin(2 * PI * 7 / 12)
    8 : Août=> cos(2 * PI * 7 / 7) & sin(2 * PI * 8 / 12)
    9 : Septembre=> cos(2 * PI * 7 / 7) & sin(2 * PI * 9 / 12)
    10 : Octobre=> cos(2 * PI * 7 / 7) & sin(2 * PI * 10 / 12)
    11 : Novembre=> cos(2 * PI * 7 / 7) & sin(2 * PI * 11 / 12)
    12 : Décembre=> cos(2 * PI * 7 / 7) & sin(2 * PI * 12 / 12)
    
Ajout du numéro du trimestre Nous allons le faire comme suit : 

    1 : 1er trimestre => cos(2 * PI * 1 / 4) & sin(2 * PI * 1 / 4)
    2 : 2ème trimestre => cos(2 * PI * 2 / 4) & sin(2 * PI * 2 / 4)
    3 : 3ème trimestre => cos(2 * PI * 3 / 4) & sin(2 * PI * 3 / 4)
    4 : 4ème trimestre => cos(2 * PI * 4 / 4) & sin(2 * PI * 4 / 4)


```python
import math

def root_time(date,time_type):
    if time_type=='day':
        root_time=date.weekday()+1
        nb=7
    if time_type=='month':
        root_time=date.month
        nb=12
    if time_type=='quarter':
        root_time=math.ceil(date.month/3.)
        nb=4
    return root_time,nb

def add_columns_time(df_global, time_type,start_date,end_date):
    
    daterange = pd.date_range(start_date, end_date)
    
    root=time_type.capitalize()
    
    df_global[root+'_Cos'] = 0
    df_global[root+'_Sin'] = 0
        
    for date_ref in daterange:
        df_global[root+'_Cos'][(df_global['GOLD_Time'] == date_ref)] = math.cos(2 * math.pi * (root_time(date_ref,time_type)[0]) / root_time(date_ref,time_type)[1])
        df_global[root+'_Sin'][(df_global['GOLD_Time'] == date_ref)] = math.sin(2 * math.pi * (root_time(date_ref,time_type)[0]) / root_time(date_ref,time_type)[1])
    
    return df_global
df_global=add_columns_time(df_global,'day','01/02/2013', '31/12/2019')
df_global=add_columns_time(df_global,'month','01/02/2013', '31/12/2019')       
df_global=add_columns_time(df_global,'quarter','01/02/2013', '31/12/2019')
```

```python
def transform_df_variation(df):
    df2 = df.copy()
    for col_var in df.columns:
        #print(col_var, len(df))
        if ((col_var == 'GOLD_Time') or (col_var[-3:] == 'Cos') or (col_var[-3:] == 'Sin') or (col_var[-6:] == 'Volume') or (col_var[-3:] == 'Cap')):
            print(col_var, ': on ne fait rien sur cette colonne')
        elif (col_var[-8:] == 'Intraday'):
            taille = len(col_var) - 9
            crypto = col_var[:taille]
            col_ref = crypto + '_Low'
            df2[col_var] = df[col_var] / df[col_ref].replace(0, 1)
        else:
            for ii in range(len(df)):
                j=len(df)-ii-1
                print('\n', df[col_var].values)
                if ii == 0:
                #    print('ii = 0')
                    df2[col_var].values[j] = 0
                else:
                    if df[col_var].values[j+1] != 0:
                        df2[col_var].values[j] = (df[col_var].values[j] - df[col_var].values[j+1])/df[col_var].values[j+1]
                    else:
                        print(col_var, ': valeur = 0')
                        df2[col_var].values[j] = 0
                #print('df2 :', j, df2[col_var].values[j])
    
    return df2
    
```
